```python
lines = []
new_data = []
old_data = []
with open('Nishant_NVMain_libquntum_1B.txt') as f:
    lines = f.readlines()
```

```python
for data in lines:
    aux = list(data.split(' '))
    aux.pop(0)
    new_data.append(aux[:16])
    old_data.append(aux[-16:])
```

```python
for i in old_data:
    i[-1] = i[-1].strip('\n')
```

```python
neww = new_data
oldd = old_data
```

```python
new_data = []
old_data = []
for i in range(len(neww)):
    flag = True
    for j in range(16):
        if oldd[i][j] != '0':
            flag = False
    if flag == True:
        continue
    if neww[i] == oldd[i]:
        continue
    else:
        new_data.append(neww[i])
        old_data.append(oldd[i])
```

```python
binary_block_new = []
for block in new_data:
    bit_block = ""
    for word in block:
        nibble_bit = ""
        for nibble in word:
            nibble_bit = nibble_bit + "{0:04b}".format(int(nibble, 16))
        while(len(nibble_bit)<32):
            nibble_bit = '0' + nibble_bit
        bit_block = bit_block + nibble_bit
    binary_block_new.append(bit_block)
```

```python
new_data[0]
```

```python
binary_block_old = []
for block in old_data:
    bit_block = ""
    for word in block:
        nibble_bit = ""
        for nibble in word:
            nibble_bit = nibble_bit + "{0:04b}".format(int(nibble, 16))
        while(len(nibble_bit)<32):
            nibble_bit = '0' + nibble_bit
        bit_block = bit_block + nibble_bit
    binary_block_old.append(bit_block)
```

```python
bitflip_block = []
for i in range(len(binary_block_new)):
    aux = ""
    for j in range(len(binary_block_new[i])):
        bit = int(binary_block_new[i][j])^int(binary_block_old[i][j])
        aux = aux + str(bit)
    bitflip_block.append(aux)
```

```python
bitflips = 0
for block in bitflip_block:
    for i in range(len(block)):
        if block[i] == '1':
            bitflips = bitflips + 1
```
