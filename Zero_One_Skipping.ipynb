```python
binary_block_new = []
for block in new_data:
    bit_block = ""
    for word in block:
        nibble_bit = ""
        for nibble in word:
            nibble_bit = nibble_bit + "{0:04b}".format(int(nibble, 16))
        while(len(nibble_bit)<32):
            nibble_bit = '0' + nibble_bit
        bit_block = bit_block + nibble_bit
    binary_block_new.append(bit_block)
```

```python
binary_block_old = []
for block in old_data:
    bit_block = ""
    for word in block:
        nibble_bit = ""
        for nibble in word:
            nibble_bit = nibble_bit + "{0:04b}".format(int(nibble, 16))
        while(len(nibble_bit)<32):
            nibble_bit = '0' + nibble_bit
        bit_block = bit_block + nibble_bit
    binary_block_old.append(bit_block)
```

```python
def getMaxLength(block, num):
    count = 0
    result = 0
    for i in range(len(block)):
        if (block[i] == num):
            count = 0
        else:
            count+= 1
            result = max(result, count)
    return result
```

```python
count_zero_new = []
count_zero_old = []
count_one_new = []
count_one_old = []
for i in range(len(binary_block_new)):
    count_zero_new.append(getMaxLength(binary_block_new[i], '1'))
    count_zero_old.append(getMaxLength(binary_block_old[i], '1'))
    count_one_new.append(getMaxLength(binary_block_new[i], '0'))
    count_one_old.append(getMaxLength(binary_block_old[i], '0'))
```
